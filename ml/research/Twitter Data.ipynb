```python
# imports 
import pymongo
import requests
import json 

# Twitter Keys 
API_KEY = ""
API_SECRET = ""
BEARER_TOKEN = ""

# Reddit Keys 
CLIENT_ID = ""
SECRET_ID = ""

# Initialise the Reddit Authentication 
auth = requests.auth.HTTPBasicAuth(CLIENT_ID, SECRET_ID)
data = {'grant_type': 'password',
        'username': '',
        'password': ''}

headers = {'User-Agent': 'Hype-Calc/0.0.1'}
res = requests.post('https://www.reddit.com/api/v1/access_token',
                    auth=auth, data=data, headers=headers)
TOKEN = res.json()['access_token']
headers = {**headers, **{'Authorization': f"bearer {TOKEN}"}}
```

```python
MONGO_STRING = ""
CLIENT = pymongo.MongoClient(MONGO_STRING)
collection = CLIENT.CollectionDB.collections
collection_names = [doc["name"] for doc in collection.find({})]
```

```python
def twitter_followers(collection):
    url = "https://api.twitter.com/1.1/users/show.json"
    querystring = {"screen_name": collection}
    headers = {"Authorization": "Bearer "+BEARER_TOKEN}
    response = requests.request(
        "GET", url, headers=headers, params=querystring)
    resp = json.loads(response.text)
    if "followers_count" in resp:
        return int(resp["followers_count"])
    return 0
```

```python
def get_score_from_reddit(collection_name):
    params = {'query': collection_name}
    res = requests.post('https://oauth.reddit.com/api/search_subreddits', headers=headers, params=params)
    if res.status_code != 200: return 0
    
    subreddits = [item['name'] for item in res.json()['subreddits']]
    total = 0
    for subreddit in subreddits:
        url = 'https://oauth.reddit.com/r/' + subreddit + '/about.json'
        res = requests.get(url, headers=headers)
        total += int(res.json()['data']['subscribers'])
    return total
```
